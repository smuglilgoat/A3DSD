```python
# Use in Blender
import bpy

for obj in bpy.data.objects:
    mesh = obj.data
    histogram = {}
    for vert in mesh.vertices:
        index = vert.index
        n_neighbours = len([e.vertices[e.vertices[0] == index] for e in mesh.edges if index in e.vertices])
        if n_neighbours in histogram:
            histogram[n_neighbours] += 1
        else:
            histogram[n_neighbours] = 1
    print(obj.name)
    print( histogram)
```
